```python
from collections import defaultdict, deque

class Candidate:
    def __init__(self, roll_no, preferences, advanced_rank):
        self.roll_no = roll_no
        self.preferences = preferences  # List of tuples (InstCode, Brcode)
        self.advanced_rank = advanced_rank  # Advanced rank of the candidate
        self.current_index = 0  # Pointer to current preference

class Program:
    def __init__(self, inst_code, brcode, capacity):
        self.inst_code = inst_code
        self.brcode = brcode
        self.capacity = capacity
        self.waitlist = []  # Will store tuples (roll_no, rank), sorted by rank

# Example dictionaries for demonstration
candidates = {
    '230310481885': Candidate('230310481885', [('101', '4109'), ('102', '4110')], 5),
    '230310127660': Candidate('230310127660', [('102', '4110'), ('101', '4109')], 1)
}

programs = {
    ('101', '4109'): Program('101', '4109', 10),
    ('102', '4110'): Program('102', '4110', 5)
}

# Create an empty queue of candidates, sorted by advanced rank
queue = deque(sorted(candidates.values(), key=lambda c: c.advanced_rank))

def process_candidates(programs, candidates, queue):
    allocations = {}

    while queue:
        candidate = queue.popleft()
        if candidate.current_index < len(candidate.preferences):
            inst_code, brcode = candidate.preferences[candidate.current_index]
            program_key = (inst_code, brcode)
            program = programs.get(program_key)

            if program:
                # Insert candidate into waitlist in sorted order by rank
                inserted = False
                for i, other in enumerate(program.waitlist):
                    if candidates[other[0]].advanced_rank > candidate.advanced_rank:
                        program.waitlist.insert(i, (candidate.roll_no, candidate.advanced_rank))
                        inserted = True
                        break
                if not inserted:
                    program.waitlist.append((candidate.roll_no, candidate.advanced_rank))

                allocations[candidate.roll_no] = program_key

                # Adjust waitlist if over capacity
                if len(program.waitlist) > program.capacity:
                    # Remove the candidate with the highest rank (last in the list)
                    removed_roll_no, _ = program.waitlist.pop()
                    allocations.pop(removed_roll_no, None)
                    candidates[removed_roll_no].current_index += 1
                    queue.append(candidates[removed_roll_no])
        else:
            print(f"Candidate {candidate.roll_no} has exhausted their preferences.")

    return allocations, programs

allocations, final_programs = process_candidates(programs, candidates, queue)

# Printing the results
print("Final Allocations:")
for roll_no, program_key in allocations.items():
    print(f"Candidate {roll_no} is allocated to program {program_key}")

print("\nFinal Waitlists:")
for program_key, program in final_programs.items():
    print(f"Program {program_key} waitlist: {[roll for roll, _ in program.waitlist]} with {program.capacity} remaining seats")
```

```python
from collections import defaultdict, deque
import heapq

class Candidate:
    def __init__(self, roll_no, preferences, advanced_rank):
        self.roll_no = roll_no
        self.preferences = preferences  # List of tuples (InstCode, Brcode)
        self.advanced_rank = advanced_rank  # Advanced rank of the candidate
        self.current_index = 0  # Pointer to current preference

class Program:
    def __init__(self, inst_code, brcode, capacity):
        self.inst_code = inst_code
        self.brcode = brcode
        self.capacity = capacity
        self.waitlist = []  # Will store tuples (roll_no, rank), sorted by rank
    
    def add_candidate(self, candidate):
        # Insert candidate into the heap as a tuple of (-rank, roll_no)
        heapq.heappush(self.waitlist, (-candidate.advanced_rank, candidate.roll_no))

    def remove_least_preferred(self):
        # Remove the candidate with the highest positive rank (least preferred)
        return heapq.heappop(self.waitlist)
        

# Example dictionaries for demonstration
candidates = {
    '230310481885': Candidate('230310481885', [('101', '4109'), ('102', '4110')], 5),
    '230310127660': Candidate('230310127660', [('102', '4110'), ('101', '4109')], 1)
}

programs = {
    ('101', '4109'): Program('101', '4109', 10),
    ('102', '4110'): Program('102', '4110', 5)
}

# Create an empty queue of candidates, sorted by advanced rank
queue = deque(sorted(candidates.values(), key=lambda c: c.advanced_rank))

def process_candidates(programs, queue):
    allocations = {}

    while queue:
        candidate = queue.popleft()
        if candidate.current_index < len(candidate.preferences):
            inst_code, brcode = candidate.preferences[candidate.current_index]
            program_key = (inst_code, brcode)
            program = programs.get(program_key)

            if program:
                # Add candidate to the program's waitlist (max heap)
                program.add_candidate(candidate)
                allocations[candidate.roll_no] = program_key

                # Ensure capacity limits are enforced
                if len(program.waitlist) > program.capacity:
                    # Pop the least preferred candidate
                    _, removed_roll_no = program.remove_least_preferred()
                    allocations.pop(removed_roll_no, None)
                    candidates[removed_roll_no].current_index += 1
                    queue.append(candidates[removed_roll_no])
        else:
            print(f"Candidate {candidate.roll_no} has exhausted their preferences.")

    return allocations, programs


allocations, final_programs = process_candidates(programs, queue)

# Printing the results
print("Final Allocations:")
for roll_no, program_key in allocations.items():
    print(f"Candidate {roll_no} is allocated to program {program_key}")

print("\nFinal Waitlists:")
for program_key, program in final_programs.items():
    # Convert max-heap to a sorted list of roll numbers
    sorted_waitlist = sorted(program.waitlist, key=lambda x: x[0], reverse=True)
    print(f"Program {program_key} waitlist: {[roll for _, roll in sorted_waitlist]} with {program.capacity} remaining seats")
```

```python
import pandas as pd
from collections import deque
import heapq
import time

class Program:
    def __init__(self, inst_code, brcode, capacity, min_cutoff=0, round_no=1):
        self.inst_code = inst_code
        self.brcode = brcode
        self.capacity = capacity
        self.min_cutoff = min_cutoff
        self.round_no = round_no
        self.waitlist = []

    def is_eligible(self, candidate):
        if self.round_no ==1 :
            return True
        return candidate.adv <= self.min_cutoff

    def add_candidate(self, candidate):
        # Only add if eligible by rank
        if self.is_eligible(candidate):
            heapq.heappush(self.waitlist, (-candidate.adv, candidate.roll_no))
            return True
        return False

    def remove_least_preferred(self):
        # Remove the candidate with the highest positive rank (least preferred)
        return heapq.heappop(self.waitlist)

class Candidate:
    def __init__(self, roll_no, adv_rank, mains_rank):
        self.roll_no = roll_no
        self.adv = adv_rank
        self.mains = mains_rank
        self.current_index = 1  # Initialize to 1 assuming choice numbers start from 1

# Load data from CSV files
def load_candidates(filename):
    df = pd.read_csv(filename)
    filtered_df = df[(df['Adv_CRL_Rank'] > 0) & 
                     (df['Gender'] == 'B') & 
                     (df['Cat'] == 'OP') & 
                     (df['Adv_DS'] == 2) & 
                     (df['Nationality'] == 1) & 
                     (df['SubCat'] == 'NO')].sort_values('Adv_CRL_Rank').head(100)
    return {row['RollNo']: Candidate(row['RollNo'], row['Adv_CRL_Rank'], row['AI_Eng_CRL_Rank']) for index, row in filtered_df.iterrows()}


def load_programs(filename):
    df = pd.read_csv(filename)
    filtered_df = df[ 
                     (df['Gender'] == 'B') & 
                     (df['Cat'] == 'OP') & 
                     (df['SubCat'] == 'NO')]
    return {(row['InstCode'], row['Brcode']): Program(row['InstCode'], row['Brcode'], row['TotalSeat']) for index, row in filtered_df.iterrows()}

def load_choices(filename):
    df = pd.read_csv(filename)
    return df

# Example file paths
candidates = load_candidates(r'C:\Users\student\Desktop\JOSAA\data\Round1-20240508T102918Z-001\Round1\Candidate.csv')
programs = load_programs(r'C:\Users\student\Desktop\JOSAA\data\Round1-20240508T102918Z-001\Round1\Seats.csv')
choices = load_choices(r'C:\Users\student\Desktop\JOSAA\data\Round1-20240508T102918Z-001\Round1\Choice.csv')

# Function to get preferences based on Choices DataFrame
def get_preference(roll_no, choice_no, choices_df):
    row = choices_df[(choices_df['RollNo'] == roll_no) & (choices_df['ChoiceNo'] == choice_no)]
    if not row.empty:
        inst_code, br_code = row.iloc[0]['InstCode'], row.iloc[0]['Brcode']
        return programs.get((inst_code, br_code))
    return None

# Processing candidates
queue = deque(sorted(candidates.values(), key=lambda c: c.adv))  # Sort by advanced rank for processing
# queue = deque([candidates[roll_no] for roll_no in candidates])

def process_candidates():
    start_time = time.time()
    allocations = {}
    while queue:
        candidate = queue.popleft()
        preference = get_preference(candidate.roll_no, candidate.current_index, choices)
        if preference:
            if preference.add_candidate(candidate):
                allocations[candidate.roll_no] = (preference.inst_code, preference.brcode)
                if len(preference.waitlist) > preference.capacity:
                    last_rank = -preference.waitlist[0][0]  # Assuming waitlist stores (-rank, roll_no)
                    if last_rank > preference.min_cutoff:
                        # Count how many have the same rank as the last rank
                        count_same_rank = sum(1 for x in preference.waitlist if -x[0] == last_rank)

                        # Check if removing these candidates brings us below capacity
                        if len(preference.waitlist) - count_same_rank >= preference.capacity:
                            removed_candidates = []

                            # Remove all candidates with the least preferred rank
                            while preference.waitlist and -preference.waitlist[0][0] == last_rank:
                                _, roll_no = heapq.heappop(preference.waitlist)  # Pop the least preferred candidate
                                removed_candidates.append(roll_no)

                            # Requeue the removed candidates for further processing
                            for roll_no in removed_candidates:
                                candidates[roll_no].current_index += 1  # Move to next preference
                                candidate = candidates[roll_no] 
                                queue.append(candidate)  # Requeue for another attempt
            else:
                candidates[candidate.roll_no].current_index += 1
                queue.append(candidate)  # Add back to the queue if not placed  
                
    print(f"Time taken: {time.time() - start_time:.2f} seconds")
    return allocations, programs

allocations , final_programs = process_candidates()


print("Final Allocations:")
for roll_no, (inst_code, brcode) in allocations.items():
    candidate = candidates.get(roll_no)
    print(f"Candidate {roll_no} Rank - {candidate.adv}  Choice - {candidate.current_index} - is allocated to program {inst_code}'s {brcode}")
    
    



print("\nFinal Waitlists:")
for program_key, program in final_programs.items():
    if program.waitlist:
        print(f"Program {program_key} waitlist: {program.waitlist} with {program.capacity - len(program.waitlist)} remaining seats")
    
    
print("\nFinal Cutoffs:")
for program_key, program in final_programs.items():
    if program.waitlist:
        last_rank = -program.waitlist[0][0]
        print(f"Program {program_key} cutoff: {last_rank}")
    # else:
    #     print(f"Program {program_key} has no waiting list")
```
